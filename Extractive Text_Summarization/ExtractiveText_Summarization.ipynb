```python
#Text Summarization
```

```python
print(text.count("."))
```

```python
# Remove punctuation
```

```python
string.punctuation = string.punctuation + '\n' 
print(string.punctuation)
```

```python
nopunct=[char for char in text if char not in string.punctuation]
nopunct="".join(nopunct)
nopunct
```

```python
# Remove stopwords
```

```python
process_text=[word for word in nopunct.split() if word.lower() not in nltk.corpus.stopwords.words()]
process_text
```

```python
word_freq={}
for word in process_text:
  if word not in word_freq:
    word_freq[word]=1
  else:
      word_freq[word]=word_freq[word]+1

dict(sorted(word_freq.items(),key=lambda item:item[1],reverse=True))


```

```python
max_freq=max(word_freq.values())

for word in word_freq.keys():
  word_freq[word]=(word_freq[word]/max_freq)
word_freq
```

```python
# Tokenization
```

```python
sent_list=nltk.sent_tokenize(text)
sent_list
    
```

```python
sent_score={}

for sent in sent_list:
  for word in nltk.word_tokenize(sent.lower()):
    if word in word_freq.keys():
       if sent not in sent_score.keys():
         sent_score[sent]=word_freq[word]
       else:
         sent_score[sent]=sent_score[sent]+word_freq[word]

dict(sorted(sent_score.items(),key=lambda item:item[1],reverse=True))  
```

```python
summary_sent=nlargest(3,sent_score, key=sent_score.get)

summary_sent
```
